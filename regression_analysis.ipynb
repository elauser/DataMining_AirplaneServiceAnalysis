## Multiple Linear Regression Model1

```python
from sklearn import linear_model
import statsmodels.api as sm

y_rating = df.overall_rating
y_recommended = df.recommended

x = df.drop(columns=['overall_rating', 'recommended', 'cabin_flown', 'positive', 'negative'])
x_model1 = x.iloc[:,:5]
x_model2 = x.iloc[:,:14]
x_model3 = x
```

### Model1

```python
est = sm.OLS(y_rating, x_model1)
est2 = est.fit()
est2.summary()
```

### Model2

```python
est = sm.OLS(y_rating, x_model2)
est2 = est.fit()
est2.summary()
```

### Model3

```python
est = sm.OLS(y_rating, x_model3)
est2 = est.fit()
est2.summary()
```
